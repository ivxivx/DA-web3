```python
import os

flipsideApiKey = os.environ.get("FLIPSIDE_API_KEY")
```

```python
import pandas as pd
from flipside import Flipside

flipside = Flipside (flipsideApiKey, "https://api-v2.flipsidecrypto.xyz")

sql = """SELECT * from ethereum.core.fact_transactions limit 1"""

query_result_set = flipside.query (sql)
transaction_df = pd.DataFrame(query_result_set.rows, columns = query_result_set.columns)
transaction_df.head()
```

Retrive the latest block

```python
sql = """SELECT * from ethereum.core.fact_transactions order by block_number desc limit 1"""

query_result_set = flipside.query (sql)
transaction_df = pd.DataFrame(query_result_set.rows, columns = query_result_set.columns)
transaction_df.head()
```
